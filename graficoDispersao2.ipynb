```
import pandas as pd
import csv
import sklearn
import matplotlib.pyplot as plt
import numpy as np
```

```
df = pd.read_csv ('/content/drive/My Drive/IFPB/dataGraf2.csv')
```

```
df
```

##Gráficos dos Negativos

```
Y = df['negativo']
X = df['data']

w = 40
h = 20
d = 70
plt.figure(figsize=(w, h), dpi=d)
plt.xlim(-1, 59)

plt.scatter(X,Y, color="red", s=500, marker="x")
plt.title('Gráfico de Dispersão - Negativos', fontsize=70)

plt.xlabel('Data', fontsize=40, labelpad=20)
plt.xticks(rotation=45)
plt.ylabel('Quantidade', fontsize=40, labelpad=20)

plt.rcParams['xtick.labelsize'] = 25
plt.rcParams['ytick.labelsize'] = 30


plt.tight_layout()
plt.show()
```

```
fig, ax = plt.subplots(figsize=(50, 10))
fig.suptitle("Gráfico de Dispersão - Negativos", fontsize=50)
ax.plot(df.data,df.negativo, color="red")

#adiciona xlabel e ylabel com fontsize 30
plt.setp(ax.set_xlabel('Data'), fontsize=30)
plt.setp(ax.set_ylabel('Quantidade'), fontsize=30)

plt.setp(ax.get_xticklabels(), rotation = 45)

plt.show()
```

##Gráfico Positivos e Negativos


```
import matplotlib.ticker as ticker
```

```
fig, ax = plt.subplots(figsize=(25, 10))

#titulo do grafico
fig.suptitle("Gráfico de Dispersão", fontsize=30)

#plota o gráfico com os valores positivos
ax.plot(df.data, df.positivo, 'blue', linestyle='--', label='Positivo')
#plota o gráfico com os valores negativos
ax.plot(df.data, df.negativo, 'red', linestyle='-', label='Negativo' )

#labels de x e y 
ax.set_ylabel('Quantidade', fontsize=20)
ax.set_xlabel('Data', fontsize=20)

#legenda do gráfico
ax.legend(loc='upper right', prop={'size': 15})

#muda o tamanho dos ticks de data e quantidade
ax.tick_params(labelsize=15)

#altera o intervalo de datas que irão aparecer para 5 em 5 dias
tick_spacing = 5
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))

plt.show()

#salva o grafico em jpeg
fig.savefig('graficoDispersao.jpeg', format='jpeg')
```

```
Y = df['positivo']
X = df['data']

x1 = df['data']
y1 = df['negativo']

w = 40
h = 20
d = 70
plt.figure(figsize=(w, h), dpi=d)
plt.xlim(-1, 59)

plt.scatter(X,Y, color="blue", s=500, marker="x", label='Positivo')
plt.scatter(x1,y1, color="red", s=500, marker="x", label='Negativo')

plt.title('Gráfico de Dispersão', fontsize=70)
plt.legend(loc='upper right', prop={'size': 25})

plt.xlabel('Data', fontsize=40, labelpad=20)
plt.xticks(rotation=45)
plt.ylabel('Quantidade', fontsize=40, labelpad=20)

plt.rcParams['xtick.labelsize'] = 25
plt.rcParams['ytick.labelsize'] = 30


plt.tight_layout()
plt.show()
```
